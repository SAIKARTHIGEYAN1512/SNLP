```
# TASK 6: BASIC TEXT PROCESSING PIPELINE
import nltk
sent = "Natural language processing (NLP) is a subfield of linguistics, computer science, information engineering, and artificial intelligence concerned with the interactions between computers and human (natural) languages, in particular how to program computers to process and analyze large amounts of natural language data."
words = nltk.word_tokenize(sent)
print(words)
```

```
texts = ["""Natural language processing (NLP) is a subfield of linguistics, computer science, information engineering, and artificial intelligence concerned with the interactions 
between computers and human (natural) languages, in particular how to program computers to process and analyze large amounts of natural language data.
Challenges in natural language processing frequently involve speech recognition, natural language understanding, and natural language generation."""]

for text in texts:
    sentences = nltk.sent_tokenize(text)
    for sentence in sentences:
        words = nltk.word_tokenize(sentence)
        print(words)
        tagged = nltk.pos_tag(words)
        print(tagged)
```

```
def gender_features(word): #gives last letter of the word
    return {'last_letter':word[-1]}
gender_features('karthi')
```

```
#task 8 gender identification
from nltk.corpus import names
import random
labeled_names = ([(name, 'male') for name in names.words('male.txt')]+
                 [(name, 'female') for name in names.words('female.txt')])
random.shuffle(labeled_names)
```

```
featuresets = [(gender_features(n),gender) for (n,gender) in labeled_names]
```

```
train_set, test_set = featuresets[500:], featuresets[:500]
```

```
import nltk
classifier = nltk.NaiveBayesClassifier.train(train_set)
```

```
print(nltk.classify.accuracy(classifier, test_set))
```

```
classifier.classify(gender_features('ram'))
```

```
classifier.classify(gender_features('Sophia'))
```

```
#task 9 Vectorisers & cosine Similarity
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer(binary = True)
corpus = ["I love horror movies","Lights out is a horror movie" ]
vect.fit(corpus) 
vocab = vect.vocabulary_
```

```
#index
for key in sorted(vocab.keys()): 
     print("{}:{}".format(key, vocab[key]))
```

```
print(vect.transform(["Lights out is a horror movie"]).toarray())
```

```
print(vect.transform(corpus).toarray())
```

```
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vect.transform(["I love horror movies"]).toarray(), vect.transform(["Lights out is a horror movie"]).toarray()) 
print(similarity)
```

```
#task 10 conditional frequency distributions
from nltk.corpus import brown
cfd = nltk.ConditionalFreqDist(
              (genre, word)
              for genre in brown.categories()
              for word in brown.words(categories=genre))
```

```
genres = ['news', 'religion', 'hobbies', 'science_fiction', 'romance', 'humor']
modals = ['can', 'could', 'may', 'might', 'must', 'will']
```

```
cfd.tabulate(conditions=genres, samples=modals)
```
